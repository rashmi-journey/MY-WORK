# 1-Show the year_wise average Pollution with a Line chart 

```python
data.head()
```

```python
data['year']=data.Timestamp.dt.year
```

```python
data
```

```python
year_avg = data.groupby("year")['PM2.5'].mean()
```

```python
year_avg
```

```python
year_avg.plot(kind='line',figsize=(15,5));
```

# 2-Draw the Area plot showing the average pollution month wise 

```python
 data['Month']=data.Timestamp.dt.month
```

```python
data.head()
```

```python
data
```

```python
month_avg =data.groupby('Month')['PM2.5'].mean()
```

```python
month_avg 
```

```python
month_avg.plot(kind='area',figsize=(12,5),color='green')
```

# 3-Draw a bargraph to show the Hour wise average pollution

```python
data.head()
```

```python
data['Hour']=data.Timestamp.dt.hour
```

```python
data
```

```python
hour_avg=data.groupby('Hour')['PM2.5'].mean()
```

```python
hour_avg 
```

```python
hour_avg.plot(kind='bar',figsize=(18,5),color ='purple')
```
