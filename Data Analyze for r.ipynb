# unique()

# value_counts()

```python
A = data['churn'].value_counts()           # to show the count of unique values of the column 'churn'
print(A)                                   # ... and saving result in variable 'A' and printing variable 'A'
```

```python
str(483/3333*100) + '%'                    # Just to check the percentage of true (churned customers) from the dataset...
                                           #... and showing the result in string format , after adding % sign at the end 
```

```python
type(A)
```

# Donut Chart 

```python
plt.pie(A, labels=['Not churned' ,'Churned'],colors=['orange','green'],startangle=50,shadow=True,radius=1.5,
  explode=(0,0.2),autopct='%1.2f%%',pctdistance=0.75)
circle = plt.Circle((0,0),1,color='white')
c= plt.gcf()
c.gca().add_artist(circle)
plt.title("Churned vs Non-Churned Customers (DSL)")
plt.show()
```
