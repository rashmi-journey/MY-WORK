# Donu chart

```python
plt.pie(A,labels= ['Not churned','Churned'],colors=['lime','orange'],startangle=50,shadow=True,radius=1.7,explode=(0,0.2)
        ,autopct='%1.2f%%',pctdistance=0.75)
circle=plt.Circle((0,0),1,color='white')
c=plt.gcf()
c.gca().add_artist(circle)
plt.title("Churned vs Non-Churned Customer(DSL)")
plt.show()


```

# countplot

```python

# Setting the figure size
plt.figure(figsize=(20, 7))

# Creating the count plot for the 'state' column
sns.countplot(x='state', data=data)  # Use x='state' and specify the data argument

# Adjusting the size of x and y ticks
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

# Optional: Adding labels and title for clarity
plt.xlabel('State', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Count of States', fontsize=18)

plt.xlabel('States',fontsize=20,color='Red')
plt.ylabel('Count',fontsize=20,color='Green')

# Display the plot
plt.show()
```

```python
# Creating a count plot with hue for 'churn'
sns.set(style='whitegrid')
sns.countplot(x='state', hue='churn', data=data)  # Specify x, hue, and data arguments correctly

# Optional: Adding labels and title for clarity
plt.xlabel('State')
plt.ylabel('Count')
plt.title('State Count by Churn')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.xlabel('State',fontsize=20, color='red')
plt.ylabel('Count',fontsize=20, color='Green')

plt.figure(figsize=(20,7))

# Display the plot
plt.show()
```
