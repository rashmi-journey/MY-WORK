```python
import pandas as pd
import numpy as np
```

```python
data = data.merge(data2,on='Name')
data
```

```python
# checking the data info
data.info()
```

```python
# filling the null values

avg_age= data['Age'].mean()
data['Age']=data['Age'].fillna(avg_age)
data['Math Score']=data['Math Score'].fillna(0)
data.info()
```

```python
#unique.values

data['City'].unique()
```

```python
# replacing values and strings
data['City']=data['City'].str.replace('Port','Pt.')
data['City'].unique()
```

```python
# replacing values and value counts
data['Gender'] = data['Gender'].replace({'Male':'M','Female':'F','other':'O'})
```

```python
data['Gender'].value_counts(normalize=True) 
```

```python
# group by function 
# data.groupby('Gender')['Math Score'].sum()
data.groupby('Gender')['Math Score'].agg(['sum','mean'])
```

```python
# pivot table
import numpy as np
pd.pivot_table(data,values =['Math Score','Science Score','English Score'],index ='Gender',aggfunc=np.mean,margins=True)
```

```python
#transform 
data['Total Score']=data['Math Score']+ data['English Score']+ data['Science Score']
data['Average per Group'] = data.groupby('Gender')['Total Score'].transform('mean')
data.head()
```

```python
#numpy conditional

avg_math= data['Math Score'].mean()
data['Average_math_score']=np.where(data['Math Score']>avg_math,'Yes','No')
data['Average_math_score'].value_counts(normalize=True)
```

```python
# function creation 
def above_avg(df, col, new_column):
    col_mean = df[col].mean()  # Calculating the mean of the specified column
    df[new_column] = np.where(df[col] > col_mean, 'Yes', 'No')  # Creating a new column with 'Yes' if value is above the mean
    return df
```

```python
above_avg(data, 'Science Score', 'Science Score Above Avg')
above_avg(data, 'Science Score', 'English Score Above Avg')
```
